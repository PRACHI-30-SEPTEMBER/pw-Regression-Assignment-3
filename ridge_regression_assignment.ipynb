# Ridge Regression Assignment

## Questions and Answers

## Q1. What is Ridge Regression?

**Answer:**
Ridge Regression is a regularization technique that adds an L2 penalty term to reduce overfitting.

Formula:
Loss = RSS + λ∑β²

Difference from OLS:
- OLS minimizes only error
- Ridge adds penalty and shrinks coefficients


## Q2. Assumptions of Ridge Regression

**Answer:**
- Linearity
- Independence of errors
- Homoscedasticity
- Normal distribution of errors
- No perfect multicollinearity


## Q3. Selecting Lambda

**Answer:**
- Cross-validation
- Grid search
- Choose value with minimum error


## Q4. Feature Selection

**Answer:**
Ridge does not perform feature selection.
It only reduces coefficient values but does not make them zero.


## Q5. Multicollinearity

**Answer:**
Ridge handles multicollinearity by stabilizing coefficients and reducing variance.


## Q6. Categorical & Continuous Variables

**Answer:**
Yes:
- Continuous → direct use
- Categorical → use encoding (One-hot encoding)


## Q7. Interpreting Coefficients

**Answer:**
- Shows relationship strength
- Values shrink compared to OLS
- Interpretation remains same


## Q8. Ridge in Time Series

**Answer:**
Yes:
- Use lag features
- Apply regression on time-based variables


## Practical Implementation

```
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100, n_features=5, noise=10)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = Ridge(alpha=1.0)
model.fit(X_train, y_train)

print("Model Score:", model.score(X_test, y_test))
```
